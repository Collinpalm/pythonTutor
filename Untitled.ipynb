```python
import numpy as np
import pandas as pd
from pandas import read_csv


#read in data from csv using pandas
df = pd.read_csv('data.csv', sep=',',header=None)
#save data to variable
data = df.values
#save data to numpy array
narr = np.array(data)

print(narr)
for i in range(0, (narr.size-2)):
    print(i)
    print("Min: "+ str(np.min(narr[i])))
    print("Max: "+ str(np.max(narr[i])))
    print("Med: "+ str(np.median(narr[i])))
    print("Mean: "+ str(np.mean(narr[i])))
    print("Std: "+ str(np.std(narr[i])))
    
    
```
